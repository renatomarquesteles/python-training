```python
import pyautogui
import time

# wait 0.5s after each pyautogui instruction
pyautogui.PAUSE = 0.5

# Step 1: Open company's website (link)

pyautogui.hotkey("ctrl", "t") # open a new browser tab
pyautogui.write("https://pages.hashtagtreinamentos.com/aula1-intensivao-sistema")
pyautogui.press("enter")

# wait for the page to load
time.sleep(2)

# Step 2: Log-in

pyautogui.click(x=947, y=350) # select login input
pyautogui.write("my_login")
pyautogui.press("tab")
pyautogui.write("my_password")
pyautogui.click(x=957, y=493) # click login button

# wait for the page to load
time.sleep(3)

# Step 3: Export database
pyautogui.click(x=885, y=267, button="right") # right-click on file
pyautogui.press("up") # select download button
pyautogui.press("enter")

# wait for file download
time.sleep(4)

```

```python
import pandas as pd # rename pandas to pd

# Step 4: Calculate indicators
table = pd.read_csv(r"C:\Users\PC\Downloads\Compras.csv", sep=";") # r"string" -> raw string
display(table)

total_spent = table["ValorFinal"].sum()
quantity = table["Quantidade"].sum()
avg_price = total_spent / quantity

print(total_spent)
print(quantity)
print(avg_price)

```

```python
import pyperclip

# Step 5: Send email to boss
pyautogui.hotkey("ctrl", "t") # open a new browser tab
pyautogui.write("https://mail.google.com/mail/u/0/#inbox")
pyautogui.press("enter")

time.sleep(3)

pyautogui.click(x=78, y=170) # click Compose button
pyautogui.write("renatomarquesteles@gmail.com")
pyautogui.press("tab") # select recipient
pyautogui.press("tab") # jumps to subject input

pyperclip.copy("Sales Report - Márçh, 2023") # "hack" to allow special characters writing
pyautogui.hotkey("ctrl", "v")

pyautogui.press("tab") # jumps to email body

body = f"""Dear Boss,

Here is the sales report for this month

Total spent: ${total_spent:,.2f}
Number of products purchased: ${quantity:,}
Average price: ${avg_price:,.2f}

Best regards,
Renato
"""
# f"string" -> format string

pyautogui.write(body)
pyautogui.hotkey("ctrl", "enter") # sends email


```
